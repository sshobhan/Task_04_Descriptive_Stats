## Script 1 — Pure Python (No Pandas, No Polars)


```python
import csv

# Open the file
with open("C:\\Users\\s29sh\\OneDrive\\Documents\\Datasets\\period_03\\2024_tw_posts_president_scored_anon.csv", mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    header = next(reader)
    first_row = next(reader)

print("Header:\n", header)
print("\nFirst Row:\n", first_row)
```

```python
##2. Descriptive Statistics for Numeric Columns
import csv
import statistics

csv_file = "2024_tw_posts_president_scored_anon.csv"

metrics = ["retweetCount", "replyCount", "likeCount", "quoteCount", "viewCount"]

metrics_data = {metric: [] for metric in metrics}

with open(csv_file, encoding="utf-8") as f:
    tw_reader = csv.DictReader(f)
    for tweet in tw_reader:
        for metric in metrics:
            val = tweet[metric]
            try:
                metrics_data[metric].append(int(val))
            except (ValueError, TypeError):
                continue  # Skip if not convertible to int

def summary_stats(nums):
    if not nums:
        return None
    return {
        "Count": len(nums),
        "Mean": round(statistics.mean(nums), 2),
        "Min": min(nums),
        "Max": max(nums),
        "Std Dev": round(statistics.stdev(nums), 2) if len(nums) > 1 else 0.0
    }

for metric in metrics:
    print(f"Summary for '{metric}':")
    print(summary_stats(metrics_data[metric]))
    print("-" * 40)
```

```python
#3. Frequency Counts for Categorical Variables
from collections import Counter

# Categorical columns to analyze
categorical_columns = ["lang", "isReply", "isRetweet", "source"]

# Store frequency counts
category_counts = {col: Counter() for col in categorical_columns}

# Read and count values
with open(csv_file, mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        for col in categorical_columns:
            value = row[col].strip()
            category_counts[col][value] += 1

# Display results
for col in categorical_columns:
    print(f"Column: {col}")
    print(f"Unique values: {len(category_counts[col])}")
    print("Top 5 most common values:")
    for value, count in category_counts[col].most_common(5):
        print(f"  {value}: {count}")
    print("-" * 40)
```

```python
#4. Grouped Aggregations by Language

import csv
import statistics

group_by = "lang"
summarize_fields = ["likeCount", "viewCount"]

# Grouping storage
language_metrics = {}

with open(csv_file, encoding="utf-8") as f:
    tweet_reader = csv.DictReader(f)
    for tweet in tweet_reader:
        lang_key = tweet[group_by].strip()
        if lang_key not in language_metrics:
            language_metrics[lang_key] = {field: [] for field in summarize_fields}
        for field in summarize_fields:
            try:
                language_metrics[lang_key][field].append(int(tweet[field]))
            except (ValueError, TypeError):
                continue

# Print group summaries
for lang, field_data in language_metrics.items():
    print(f"Language Group: {lang}")
    for field in summarize_fields:
        vals = field_data[field]
        if vals:
            avg = round(statistics.mean(vals), 2)
            cnt = len(vals)
            print(f"  {field} → Mean: {avg}, Count: {cnt}")
    print("-" * 40)
```

## Insights from Output:
The majority of tweets (27,281) were in English (en), making it the most prevalent language in the dataset.

Tagalog (tl) tweets showed a notably high average number of likes (approximately 23,366), but this figure is based on just 4 tweets, suggesting these numbers may be skewed by outliers.

Tweets in other languages, such as Haitian Creole (ht) and Turkish (tr), recorded exceptionally high average view counts, which is likely a result of the small number of tweets in those languages.

## Script 3 — With Polars

```python
# Load the CSV into a Polars DataFrame
df_pl = pl.read_csv("2024_tw_posts_president_scored_anon.csv")

# Show shape and preview
print("Shape:", df_pl.shape)
df_pl.head()
```

```python
# Descriptive statistics for numeric columns
df_pl.describe()
```

```python
# Categorical columns to inspect
cat_cols = ["lang", "source", "isReply", "isRetweet"]

for col in cat_cols:
    print(f"\nColumn: {col}")
    value_counts = (
        df_pl
        .group_by(col) 
        .agg(pl.count())
        .sort("count", descending=True)
        .head(5)
    )
    print(value_counts)
    print("-" * 40)
```

```python
# Grouped mean and count of likeCount and viewCount by language
grouped_stats = (
    df_pl
    .group_by("lang")
    .agg([
        pl.col("likeCount").mean().alias("likeCount_mean"),
        pl.col("likeCount").count().alias("likeCount_count"),
        pl.col("viewCount").mean().alias("viewCount_mean"),
        pl.col("viewCount").count().alias("viewCount_count")
    ])
    .sort("likeCount_count", descending=True)
)
print(grouped_stats)
```

## Insights from Output:
We have obtained the mean and total count for both likeCount and viewCount, grouped by each language (lang).

The results are organized so that languages with the highest number of likeCount entries appear at the top.

This approach aligns precisely with the Pandas-based solution from Script 2.
